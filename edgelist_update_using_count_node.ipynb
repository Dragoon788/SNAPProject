```python
import pandas as pd
group_ds = pd.read_csv("channels_with_count.csv")
og_github_edgelist = pd.read_csv("10.05.2021.13.04EDGELIST.csv")
```

```python
group_ds

# Merging Source column with group_ds
edgelist_with_source = pd.merge(og_github_edgelist, group_ds[['Channel_Name']], left_on='Source', right_on='Channel_Name', how='left')
edgelist_with_source

# Merge Target column with group_ds
edgelist_with_both = pd.merge(edgelist_with_source, group_ds[['Channel_Name']], left_on='Target', right_on='Channel_Name', how='left', suffixes=('_Source', '_Target'))
edgelist_with_both

# Drop rows with NaN values in either Source or Target columns
edgelist_cleaned = edgelist_with_both.dropna(subset=['Channel_Name_Source', 'Channel_Name_Target'])
edgelist_cleaned

edgelist_cleaned = edgelist_cleaned.drop(columns= ['Channel_Name_Source', 'Channel_Name_Target'])
edgelist_cleaned
```
